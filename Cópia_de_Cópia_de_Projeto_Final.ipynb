## Perguntas elaboradas:

**1. Qual é a distribuição de idades dos pacientes?**

Este gráfico de histograma mostra a distribuição das idades dos pacientes no conjunto de dados. Isso nos permite entender a faixa etária dos pacientes e identificar se há alguma tendência ou anomalia na distribuição das idades.

> Adicionar aspas



```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], bins=30, kde=True)
plt.title('Distribuição de Idades dos Pacientes')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()
```

**2. Qual é a distribuição de gênero dos pacientes?**

Este gráfico de barras mostra a distribuição dos gêneros dos pacientes no conjunto de dados.
Ele nos permite visualizar a proporção de pacientes masculinos e femininos e identificar se há um desequilíbrio de gênero significativo.


```python
plt.figure(figsize=(8, 6))
sns.countplot(data['Gender'])
plt.title('Distribuição de Gênero dos Pacientes')
plt.xlabel('Gênero')
plt.ylabel('Contagem')
plt.show()
```

**3. Qual é a média do valor da fatura por tipo de admissão?**

Este gráfico de boxplot compara o valor das faturas para diferentes tipos de admissão.
Ele nos permite visualizar a distribuição dos valores das faturas e identificar se há diferenças significativas entre os tipos de admissão.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Admission Type', y='Billing Amount', data=data)
plt.title('Valor da Fatura por Tipo de Admissão')
plt.xlabel('Tipo de Admissão')
plt.ylabel('Valor da Fatura')
plt.show()
```

**4. Qual é a distribuição dos tipos sanguíneos dos pacientes?**

Este gráfico de barras mostra a distribuição dos tipos sanguíneos dos pacientes no conjunto de dados.
Ele nos permite visualizar a frequência de cada tipo sanguíneo e identificar

1.   Item da lista
2.   Item da lista

se há algum tipo predominante ou raro.


```python
plt.figure(figsize=(8, 6))
sns.countplot(data['Blood Type'])
plt.title('Distribuição dos Tipos Sanguíneos dos Pacientes')
plt.xlabel('Tipo Sanguíneo')
plt.ylabel('Contagem')
plt.show()
```

**5. Qual é a relação entre a condição médica e os resultados dos testes?**

 Este gráfico de barras agrupadas mostra a relação entre a condição médica e os resultados dos testes.
Ele nos permite visualizar como os diferentes resultados dos testes estão distribuídos em relação às diferentes condições médicas.

```python
plt.figure(figsize=(12, 8))
sns.countplot(x='Medical Condition', hue='Test Results', data=data)
plt.title('Condição Médica e Resultados dos Testes')
plt.xlabel('Condição Médica')
plt.ylabel('Contagem')
plt.legend(title='Resultados dos Testes')
plt.show()
```

**6. Qual é a média de idade dos pacientes por provedor de seguro?**

Este gráfico de barras mostra a média de idade dos pacientes para cada provedor de seguro.
Ele nos permite identificar se há diferenças na idade média dos pacientes entre os diferentes provedores de seguro.

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='Insurance Provider', y='Age', data=data, estimator=lambda x: sum(x) / len(x))
plt.title('Idade Média por Provedor de Seguro')
plt.xlabel('Provedor de Seguro')
plt.ylabel('Idade Média')
plt.show()
```
